````python
'''

## 유니코드 문자와 ASCII 문자의 차이점

### ASCII (American Standard Code for Information Interchange)
- **범위**: 0-127까지의 숫자
- **문자 수**: 128개 문자 (영문자, 숫자, 특수문자)
- **용량**: 1바이트로 표현
- **제한사항**: 영어만 지원, 다른 언어는 표현 불가

### 유니코드 (Unicode)
- **범위**: 0-1,114,111까지의 숫자 (UTF-8 기준)
- **문자 수**: 수만 개 이상의 문자 지원
- **용량**: 1-4바이트로 표현 (UTF-8 방식)
- **장점**: 
  - 모든 언어(한국어, 중국어, 일본어, 영어 등) 지원
  - 이모지, 특수 기호 포함
  - 글로벌 텍스트 처리 가능

### 예시
```
ASCII: A(65), a(97), 0(48)
유니코드: A(65), a(97), 0(48), 한(44032), 🌟(128142)
```

### 결론
ASCII는 단순한 영문자 집합이고, 유니코드는 세계 모든 언어와 기호를 포괄하는 포괄적인 문자 집합입니다.

'''
````

## bytes와 bytearray의 차이점

### 기본 차이점

| 특성 | bytes | bytearray |
|------|-------|-----------|
| **변경 가능성** | 불변(immutable) | 변경 가능(mutable) |
| **데이터 수정** | 불가능 | 가능 |
| **메모리 사용** | 더 적게 사용 | 더 많이 사용 |


````python
b = b'hello'
print(type(b))  # <class 'bytes'>
print(b)        # b'hello'

# bytearray 생성
ba = bytearray(b'hello')
print(type(ba)) # <class 'bytearray'>
print(ba)       # bytearray(b'hello')

# bytes는 수정 불가
try:
    b[0] = 104  # TypeError 발생
except TypeError as e:
    print(f"Error: {e}")

# bytearray는 수정 가능
ba[0] = 102  # 'f'의 ASCII 값 : 첫번째 알파벳 h를 f로 변경 시도(260520)
print(ba)    # bytearray(b'hello')
````
