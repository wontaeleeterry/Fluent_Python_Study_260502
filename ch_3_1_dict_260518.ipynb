```python
from collections import abc
# collections 모듈에서 abc(Abstract Base Classes) : 이 객체가 특정 자료형 처럼 동작하는 지 확인하는 도구

# my_dict = []        # 비어있는 리스트를 생성하면
my_dict = {}      # 비어있는 딕셔너리 생성
isinstance(my_dict, abc.Mapping)  # isinstance(객체, 타입) : 해당 객체가 특정 타입(또는 인터페이스)을 만족하는지 확인하는 함수

# my_dict는 딕셔너리이므로 Mapping에 해당 -> True
```

```python
# Sequence : 순서가 있는 자료형인가?

from collections import abc

text = "hello"
numbers = [1, 2, 3]

print(isinstance(text, abc.Sequence))
print(isinstance(numbers, abc.Sequence))
```

```python
# Iterable : 반복 가능한가?

from collections import abc

data1 = [1, 2, 3]
data2 = {"a": 1}
data3 = 100

print(isinstance(data1, abc.Iterable))
print(isinstance(data2, abc.Iterable))
print(isinstance(data3, abc.Iterable))
```

### 사용자 정의 Mapping 클래스

```python
# Config 객체

'''
__getitem__()
__iter__()
__len__()

이 3개를 구현하면 abc.Mapping 을 상속받아 "딕셔너리처럼 동작하는 클래스" 를 만들 수 있다.
'''

from collections.abc import Mapping

class Config(Mapping):      # Mapping을 상속받음 : 이 클래스는 dict처럼 동작할 것이다.
    def __init__(self, data):
        # 내부에 실제 데이터를 저장
        self._data = data
    
    def __getitem__(self, key):
        # key로 값 가져오기 (dict 처럼)
        return self._data[key]

    def __iter__(self):
        # key들을 순회할 수 있어야 함
        return iter(self._data)

    def __len__(self):
        # 항목 개수 반환
        return len(self._data)

    # 실무에서 자주 추가하는 기능
    def get_with_default(self, key, default=None):
        return self._data.get(key, default)
```

```python
# 사용 예시
 
config_data = {
    "host": "localhost",
    "port": 8080,
    "debug": True
 }

config = Config(config_data)
```

```python
# 딕셔너리처럼 사용하기
print(config["host"])
```

```python
# 반복문 사용
for key in config:
    print(key, config[key])
```

```python
# dict처럼 items() 사용해보기

for key, value in config.items():
    print(key, value)
```

```python
# 딕셔너리로 변환

config_dict = dict(config)
print(config_dict)
```

```python
# 한 줄 출력 : 디버깅용
print(list(config.items()))
```

```python
# 길이 확인
print(len(config))
```

```python
#isinstance 검사
from collections.abc import Mapping

print(isinstance(config, Mapping))
```
