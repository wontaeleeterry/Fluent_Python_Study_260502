### Cartesian Product + List Comprehension + * Unpacking print

```python
# 데카르트 곱 생성을 위한 product 함수 import

from itertools import product

# 머신러닝/딥러닝의 하이퍼 파라미터 후보 정의 : 튜플 형태로 정의
learning_rates = (0.01, 0.1)
batch_sizes = (16, 32)
optimizers = ('SGD', 'Adam')

# 데카르트곱 + 언패킹 + 지능형 리스트
# 모든 조합을 dictionary 형태로 생성
param_grid = [
    {
        "lr" : lr,        # 학습률
        "batch" : batch,  # 배치 크기
        "opt" : opt      # 옵티마이저
    }
    # 튜플 (lr, batch, opt)를 자동으로 unpacking
    for lr, batch, opt in product(learning_rates, batch_sizes, optimizers)
]

param_grid
```

```python
# Unpacking을 활용한 출력 (*)
for lr, batch, opt in product(learning_rates, batch_sizes, optimizers):
    print(*[lr, batch, opt])

# Dictionary 형태로 출력
for p in param_grid:
    print(p)
```

```python
# ============================================================
# Cartesian Product + List Comprehension + * Unpacking + print
# ============================================================

# [목표]
# 1. Cartesian Product(데카르트 곱) 이해
# 2. List Comprehension 사용
# 3. * Unpacking 사용
# 4. print() 함수에서 * 사용법 이해


# ------------------------------------------------------------
# 1. 기본 데이터 준비
# ------------------------------------------------------------

# 색상 목록
colors = ["Black", "White"]

# 사이즈 목록
sizes = ["S", "M", "L"]


# ------------------------------------------------------------
# 2. Cartesian Product 생성
# ------------------------------------------------------------

# Cartesian Product(데카르트 곱)이란?
#
# 두 집합의 모든 가능한 조합을 만드는 것
#
# 예:
# colors = [Black, White]
# sizes  = [S, M, L]
#
# 결과:
# (Black, S)
# (Black, M)
# (Black, L)
# (White, S)
# (White, M)
# (White, L)
#
# 즉:
# 모든 색상 x 모든 사이즈 조합


# ------------------------------------------------------------
# 3. List Comprehension 사용
# ------------------------------------------------------------

# 문법:
# [표현식 for 변수1 in 반복가능객체1
#          for 변수2 in 반복가능객체2]

# 아래 코드는:
# colors 의 각 color 에 대해
# sizes 의 각 size 를 순회하며
# (color, size) 튜플 생성

products = [
    (color, size)
    for color in colors
    for size in sizes
]

# 결과:
# [
#   ('Black', 'S'),
#   ('Black', 'M'),
#   ('Black', 'L'),
#   ('White', 'S'),
#   ('White', 'M'),
#   ('White', 'L')
# ]


# ------------------------------------------------------------
# 4. 일반 출력
# ------------------------------------------------------------

print("=== 일반 출력 ===")
print(products)

# 출력:
# [('Black', 'S'), ('Black', 'M'), ...]
```

```python

# ------------------------------------------------------------
# 5. * Unpacking 사용
# ------------------------------------------------------------

# * 는 iterable(리스트, 튜플 등)의 요소를
# 개별 값으로 "풀어서(unpack)" 전달한다.


print("\n=== * Unpacking 출력 ===")

# products 리스트 안의 요소들을 하나씩 풀어서 전달
print(*products)

# 내부적으로는 아래와 거의 비슷하게 동작:
#
# print(
#     ('Black', 'S'),
#     ('Black', 'M'),
#     ('Black', 'L'),
#     ('White', 'S'),
#     ('White', 'M'),
#     ('White', 'L')
# )
```

```python

# ------------------------------------------------------------
# 6. sep 옵션과 함께 사용
# ------------------------------------------------------------

print("\n=== 줄바꿈 출력 ===")

# sep="\n"
# 각 요소 사이를 줄바꿈으로 연결

print(*products, sep="\n")

# 출력:
# ('Black', 'S')
# ('Black', 'M')
# ('Black', 'L')
# ...

```

```python

# ------------------------------------------------------------
# 7. 문자열 조합 예제
# ------------------------------------------------------------

# 이번에는 튜플 대신 문자열 생성

items = [
    f"{color} / {size}"
    for color in colors
    for size in sizes
]

print("\n=== 문자열 조합 ===")
print(*items, sep="\n")

```

```python

# ------------------------------------------------------------
# 8. 실무적인 활용 예시
# ------------------------------------------------------------

# 온라인 쇼핑몰 상품 옵션 생성 예제

cpu_list = ["i5", "i7"]
ram_list = ["8GB", "16GB"]
ssd_list = ["256GB", "512GB"]

# 가능한 모든 노트북 옵션 생성
laptops = [
    (cpu, ram, ssd)
    for cpu in cpu_list
    for ram in ram_list
    for ssd in ssd_list
]

print("\n=== 노트북 옵션 조합 ===")

# 각 조합 출력
for option in laptops:
    print(option)

# 생성되는 조합 수:
# 2 x 2 x 2 = 8개

```

```python
print(*laptops[1])   # 언패킹을 이용하여 리스트 내부의 특정값을 출력
```

```python

# ------------------------------------------------------------
# 9. * Unpacking 의 핵심 개념
# ------------------------------------------------------------

numbers = [10, 20, 30]

print("\n=== * Unpacking 핵심 ===")

# 리스트 자체 출력
print(numbers)

# 요소를 풀어서 출력
print(*numbers)

# 차이:
#
# print(numbers)
# -> [10, 20, 30]
#
# print(*numbers)
# -> 10 20 30

```

```python

# ------------------------------------------------------------
# 10. 함수 호출에서도 사용 가능
# ------------------------------------------------------------

def add(a, b, c):
    return a + b + c

values = [1, 2, 3]

# 리스트를 함수 인자로 unpacking
result = add(*values)

print("\n=== 함수 인자 Unpacking ===")
print(result)

# 내부적으로는:
# add(1, 2, 3)
# 과 동일
```
