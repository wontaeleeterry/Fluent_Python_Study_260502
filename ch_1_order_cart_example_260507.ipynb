```python
class Product:
    """상품 하나를 표현하는 클래스"""

    def __init__(self, name, price, quantity=1):
        self.name = name
        self.price = price
        self.quantity = quantity

    def total_price(self):
        return self.price * self.quantity

    def __repr__(self):
        # 개발자가 객체를 확인할 때 유용한 출력
        return (
            f"Product(name='{self.name}', "
            f"price={self.price}, quantity={self.quantity})"
        )

    def __str__(self):
        # 사용자가 보기 좋은 출력
        return (
            f"{self.name} | "
            f"{self.price:,}원 x {self.quantity}개 "
            f"= {self.total_price():,}원"
        )
```

```python

class Cart:
    """쇼핑 장바구니 클래스"""

    def __init__(self, owner):
        self.owner = owner
        self.items = []

    # 상품 추가
    def add(self, product):
        self.items.append(product)

    # 총 금액 계산
    def total_amount(self):
        return sum(item.total_price() for item in self.items)

    # len(cart)
    def __len__(self):
        return len(self.items)

    # cart[index]
    def __getitem__(self, index):
        return self.items[index]

    # cart1 + cart2
    def __add__(self, other):
        new_cart = Cart(f"{self.owner} + {other.owner}")

        # 기존 상품 복사
        new_cart.items = self.items + other.items

        return new_cart

    # print(cart)
    def __str__(self):
        lines = [f"[{self.owner}님의 장바구니]"]

        for idx, item in enumerate(self.items, start=1):
            lines.append(f"{idx}. {item}")

        lines.append("-" * 40)
        lines.append(f"총 금액: {self.total_amount():,}원")

        return "\n".join(lines)

    # 개발자용 출력
    def __repr__(self):
        return f"Cart(owner='{self.owner}', items={self.items})"

```

```python

# -------------------------
# 사용 예제
# -------------------------

# 장바구니 생성
cart1 = Cart("철수")
cart1
```

```python
# 상품 추가
cart1.add(Product("노트북", 1500000, 1))
cart1.add(Product("마우스", 30000, 2))
cart1.add(Product("키보드", 120000, 1))
```

```python

# 보기 좋은 출력
print(cart1)

```

```python

# len() 사용
print("상품 종류 개수:", len(cart1))
```

```python
# 인덱싱 사용
print(cart1[0])
```

```python
# 두 번째 장바구니
cart2 = Cart("영희")
cart2.add(Product("모니터", 400000, 1))

print(cart2[0])
```

```python

# + 연산자로 장바구니 합치기
merged_cart = cart1 + cart2

print()
print("[합쳐진 장바구니]")

len(merged_cart)
```
