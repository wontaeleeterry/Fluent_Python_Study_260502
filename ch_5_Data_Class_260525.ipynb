### Ch_데이터 클래스 빌더
- 기능은 거의 없고 단지 필드를 모아 놓은 간단한 클래스 : 데이터 클래스
- 파이썬에서 데이터 클래스(dataclass)를 사용하는 가장 큰 이유는 "보일러플레이트(반복적이고 상투적인) 코드를 줄여주고, 
가독성을 극대하게 높여주기 때문"입니다.
- 기본 클래스로 데이터를 다루려면 생성자(__init__), 문자열 표현(__repr__), 객체 비교(__eq__) 등을 일일이 구현해야 하지만, 
데이터 클래스를 사용하면 이 모든 것을 자동으로 만들어줍니다.

```python
# 읿반 클래스 구현

class RegularEmployee:
    # 생성자 메서드를 직접 정의해야 합니다.
    def __init__(self, name: str, age: int, job: str):    # 보일러플레이트 (반복적이고 상투적인) 코드들 (260525)
        self.name = name
        self.age = age
        self.job = job

    # print()했을 때 예쁘게 출력되도록 값을 직접 포맷팅해야 합니다.
    def __repr__(self):
        return f"RegularEmployee(name={self.name!r}, age={self.age!r}, job={self.job!r})"

    # 객체끼리 값이 같은지 비교(==)하기 위해 비교 연산자도 구현해야 합니다.
    def __eq__(self, other):
        if not isinstance(other, RegularEmployee):
            return NotImplemented
        return self.name == other.name and self.age == other.age and self.job == other.job
```

```python
# 객체 생성 및 테스트
emp1 = RegularEmployee("파이", 25, "개발자")
emp2 = RegularEmployee("파이", 25, "개발자")

print(emp1)         # 출력: RegularEmployee(name='파이', age=25, job='개발자')
print(emp1 == emp2) # 출력: True (비교가 가능하지만, __eq__를 지우면 False가 됩니다)
```

```python
# 데이터 클래스(dataclass)로 구현했을 때
# 똑같은 기능을 데코레이터(@dataclass) 하나로 축약한 모습입니다.

from dataclasses import dataclass

# @dataclass 데코레이터를 붙이면 __init__, __repr__, __eq__ 등이 자동으로 생성됩니다. (중요)
@dataclass
class DataClassEmployee:
    name: str
    age: int
    job: str  # 변수명과 타입 힌트만 적어주면 끝입니다.
```

```python
# 객체 생성 및 테스트
emp1 = DataClassEmployee("파이", 25, "개발자")
emp2 = DataClassEmployee("파이", 25, "개발자")

# 별도의 메서드 구현 없이도 가독성 좋은 출력을 지원합니다.
print(emp1)         # 출력: DataClassEmployee(name='파이', age=25, job='개발자')

# 두 객체의 데이터가 같으므로 자동으로 True를 반환합니다.
print(emp1 == emp2) # 출력: True
```

- 코드의 간결함: self.name = name 같은 지루한 반복 작업을 완전히 없애줍니다.  : 보일러 플레이트 코드
- 가독성 향상: 클래스가 어떤 데이터를 보유하고 있는지 한눈에 파악할 수 있습니다.
- 기본 기능 기본 탑재: print()로 출력했을 때 데이터를 보기 좋게 보여주는 __repr__과, 값이 같은지 비교해 주는 __eq__가 자동으로 구현됩니다.
- 유용한 추가 옵션: @dataclass(frozen=True) 속성을 주면 값을 수정할 수 없는 '불변 객체(Immutable)'로 만들 수도 있어 데이터 안정성을 높이기 쉽습니다.
- 💡 요약하자면: > 기능을 가진 '동작 중심의 클래스'가 아니라, 단순히 **'데이터를 담아두고 전달하는 용도의 클래스'**를 만들 때는 
데이터 클래스를 쓰는 것이 무조건 이득입니다!

```python
# pp. 220
# 파이썬의 dataclass를 사용하여 위도(Latitude)와 경도(Longitude)를 
# 나타내는 불변(Immutable) 좌표 객체를 정의한 것입니다.

from dataclasses import dataclass

@dataclass(frozen=True)   # frozen=True 옵션은 한 번 생성된 좌표 객체의 값을 실수로 변경하지 못하도록 보호하는 역할을 합니다
class Coordinate:
    lat: float
    lon: float

    def __str__(self):
        ns = 'N' if self.lat >= 0 else 'S'
        we = 'E' if self.lon >= 0 else 'W'
        return f'{abs(self.lat):.1f} degree {ns}, {abs(self.lon):.1f} degree {we}'

        # __str__ 메서드를 오버라이딩하여, 숫자로 된 좌표를 
        # 사람이 읽기 좋은 방위(N, S, E, W) 형태의 문자열로 변환해 줍니다.
```

```python
# --- 실제 사용 예시 ---

# 1. 부모 클래스(Coordinate) 객체 생성 및 출력
coord = Coordinate(37.5665, 126.9780)
print("1. 기본 좌표 객체 출력:")
print(coord)  # 출력: 37.6 degree N, 127.0 degree E
print("-" * 50)
```

### 이 클래스를 상속하여 사용하는 코드 추가 작성
- Coordinate 클래스는 단순히 지도상의 좌표만을 가집니다. 이를 상속받아 "특정 장소의 이름"이나 "도시 정보"를 추가로 저장할 수 있는 확장 클래스를 만들어 보겠습니다.
- 데이터 클래스를 상속받을 때는 부모 클래스의 필드가 먼저 선언된 후 자식 클래스의 필드가 추가되는 방식으로 동작합니다.

```python
# 상속받은 자식 클래스도 데이터 클래스의 이점을 누리기 위해 @dataclass를 붙여줍니다.
# 부모가 frozen=True이면 자식 클래스도 frozen=True를 유지하는 것이 안전합니다.
@dataclass(frozen=True)
class CityLocation(Coordinate):
    city_name: str  # 자식 클래스에서 새로 추가할 데이터 필드 (도시 이름)
    country: str    # 자식 클래스에서 새로 추가할 데이터 필드 (국가 이름)

    # 부모의 __str__ 메서드를 확장하여 도시 이름과 국가 정보까지 함께 출력하도록 오버라이딩합니다.
    def __str__(self):
        # super().__str__()을 통해 부모 클래스(Coordinate)가 계산한 위경도 문자열을 그대로 가져옵니다.
        coordinate_str = super().__str__()
        return f"[{self.city_name}, {self.country}] -> {coordinate_str}"
    
```

```python
# 2. 상속받은 자식 클래스(CityLocation) 객체 생성 및 출력
# 부모의 lat, lon 인자를 먼저 적어주고, 자식의 city_name, country 인자를 이어서 적어줍니다.
seoul = CityLocation(lat=37.5665, lon=126.9780, city_name="서울", country="대한민국")
sydney = CityLocation(lat=-33.8688, lon=151.2093, city_name="시드니", country="호주")

print("2. 상속받은 도시 위치 객체 출력:")
print(seoul)   # 출력: [서울, 대한민국] -> 37.6 degree N, 127.0 degree E
print(sydney)  # 출력: [시드니, 호주] -> 33.9 degree S, 151.2 degree E
print("-" * 50)
```

### __repr__ 과 __str__ 차이점
- 파이썬을 다루다 보면 __repr__과 __str__ 모두 객체를 문자열로 바꾸는 메서드라 헷갈리기 쉽습니다. 
- __repr__은 개발자를 위한 것(디버깅 용도), __str__은 사용자를 위한 것(출력 용도)**입니다.


```python
import datetime

now = datetime.datetime.now()

# 1. __str__()의 결과 : 사용자가 읽기 좋은 깔끔한 형태
print(str(now))  
# 출력: 2026-05-25 01:32:36.123456

# 2. __repr__()의 결과 : 개발자가 이 객체의 타입과 정확한 내부 값을 알 수 있는 형태
print(repr(now)) 
# 출력: datetime.datetime(2026, 5, 25, 1, 32, 36, 123456)
```

```python
# 커스텀 클래스에서 두 메서드를 어떻게 구별해서 사용하는가 (260525)

class User:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # 개발자를 위한 디버깅용 정보 제공
    def __repr__(self):
        return f"User(name={self.name!r}, age={self.age})"

    # 사용자를 위한 보기 좋은 문장 제공
    def __str__(self):
        return f"{self.name}님({self.age}세)"

# 객체 생성
user = User("파이", 25)

# [테스트 1] print()는 __str__을 호출합니다.
print(user)        # 출력: 파이님(25세)

# [테스트 2] 리스트 같은 컨테이너 안에 담기면 내부 객체들은 __repr__로 출력됩니다.
user_list = [user]
print(user_list)   # 출력: [User(name='파이', age=25)]
```

```python
# 실무 예제
# : 공식 내장 함수인 asdict를 사용하면 데이터클래스를 즉시 Python 딕셔너리로 변환할 수 있어, 
# JSON 직렬화나 DB 삽입이 매우 편리해집니다. 아래는 데이터베이스나 외부 API 환경에서 자주 쓰이는 데이터 변환 예제입니다.

# dataclass : 데이터를 저장하는 클래스를 간단하게 만들 수 있는 기능
# asdict: dataclass 객체를 dict 형태로 변환하는 함수

from dataclasses import dataclass, asdict   
import json

# @dataclass를 사용하면 __init__() 등을 자동 생성해준다.
@dataclass
class UserProfile:
    username: str
    email: str
    is_active: bool    # 활성화 여부 : True / False
    roles: list[str]   # 사용자 권한 목록 : 문자열만 저장하는 리스트

    def to_json(self) -> str:   # 객체 -> JSON 문자열 반환 메서드
        # asdict를 이용해 dict로 만든 후 json 문자열로 변환
        return json.dumps(asdict(self), ensure_ascii=False) 

    @classmethod   # 클래스 메서드 : 객체 생성전에 호출 가능
    def from_dict(cls, data: dict):                   
        ### cls는 클래스 자신(Class)을 가리키는 매개변수 (260525)
        # Python에서 클래스 메서드(@classmethod)를 정의할 때 첫 번째 인자로 관례상 cls를 사용합니다. 
        # 일반 메서드(인스턴스 메서드)가 객체 자신을 가리키기 위해 첫 번째 인자로 self를 사용하는 것과 대칭되는 개념입니다.
        
        # 딕셔너리 데이터를 풀어(Unpacking) 객체로 생성 : 딕셔너리를 키워드 인자로 펼치기
        return cls(**data)
        # 예:
        # {"username": "abc"}
        #
        # ↓
        #
        # username="abc"
```

```python
# --- 실무 활용 예시 ---
# 1. API 응답으로 받은 JSON 딕셔너리를 데이터클래스 객체로 변환
raw_api_data = {
    "username": "python_developer",
    "email": "dev@example.com",
    "is_active": True,
    "roles": ["admin", "user"]
}

# "객체를 선언(생성)한 후 메서드를 호출한 것"이 아니라, 
# "클래스의 메서드를 먼저 호출하여 그 결과로 객체를 생성한 것"에 가깝습니다.

# 클래스 내부에 from_dict라는 깔끔한 창구를 하나 만들어 두면, 사용하는 쪽에서는 데이터가 딕셔너리든 JSON이든 신경 쓰지 않고 
# UserProfile.from_dict(raw_data) 한 줄로 깔끔하게 객체를 얻을 수 있게 됩니다. 
# 이를 디자인 패턴에서는 팩토리(Factory) 구조라고 부릅니다.

user = UserProfile.from_dict(raw_api_data)
print(f"객체 변환 확인: {user.username} (활성화 상태: {user.is_active})")
```
