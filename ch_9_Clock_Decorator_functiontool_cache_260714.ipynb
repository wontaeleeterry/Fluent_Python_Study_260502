```python
import time

def clock(func):
    def wrapper(*args, **kwargs):
        start = time.perf_counter()      # 시작 시간

        result = func(*args, **kwargs)

        end = time.perf_counter()        # 종료 시간

        elapsed = end - start

        print(f"{func.__name__} 실행시간 : {elapsed:.6f}초")

        return result

    return wrapper  
```

```python
@clock
def slow_add(a, b):
    time.sleep(2)
    return a + b

result = slow_add(3, 5)
print(result)
```

```python
'''
functools.cache는 Python 3.9부터 추가된 데커레이터로, 
함수의 이전 계산 결과를 메모리에 저장(Cache) 해두었다가 동일한 인자로 다시 호출되면 계산을 하지 않고 저장된 결과를 즉시 반환합니다.

피보나치 수열은 재귀 호출에서 같은 값을 수없이 반복 계산하기 때문에 @functools.cache를 사용하면 성능이 극적으로 향상됩니다.

'''

# 캐시를 사용하지 않은 피보나치

import time

def fibonacci(n):
    """재귀 방식의 피보나치 (캐시 없음)"""

    if n < 2:
        return n

    return fibonacci(n - 1) + fibonacci(n - 2)


start = time.perf_counter()

result = fibonacci(40)

elapsed = time.perf_counter() - start

print(f"결과 : {result}")
print(f"실행시간 : {elapsed:.4f}초")

```

```python
# functools.cache를 이용한 개선

import functools
import time


@functools.cache
def fibonacci(n):
    """재귀 방식의 피보나치 (캐시 사용)"""

    if n < 2:
        return n

    return fibonacci(n - 1) + fibonacci(n - 2)


start = time.perf_counter()

result = fibonacci(40)

elapsed = time.perf_counter() - start

print(f"결과 : {result}")
print(f"실행시간 : {elapsed:.8f}초")

# 실행 시간은 컴퓨터마다 다르지만 수천~수만 배 이상 빨라지는 경우도 흔합니다.
```

```python
# cache 내부 확인

import functools


@functools.cache
def fibonacci(n):

    if n < 2:
        return n

    return fibonacci(n-1) + fibonacci(n-2)


print(fibonacci(20))

print(fibonacci.cache_info())
```
