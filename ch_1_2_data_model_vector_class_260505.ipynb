### 간단한 2차원 벡터 클래스
- 특별 메서드를 이용해 연산을 구현하는 Vector 클래스

```python
import math

# __xxx__ 형태는 파이썬이 자동으로 호출하는 특별한 메서드
# 특별 메서드를 사용할 경우, 객체를 "숫자처럼" 다룰 수 있다.

class Vector:
    
    # 생성자
    def __init__(self, x=0, y=0):   # 생성자 init, 객체가 생성될 때 자동 실행 메서드
        self.x = x
        self.y = y
    
    # 출력
    def __repr__(self):      # print(v) 또는 v 일 때 출력 형태 정의
        return f'Vector({self.x!r}, {self.y!r})'   # 포맷 옵션 !r : repr() 적용 지시자 - 공백 여부 확인 가능
    
    # 절댓값
    def __abs__(self):       # 벡터의 길이(거리)
        return math.hypot(self.x, self.y)
    
    # 참/거짓
    def __bool__(self):      # 참/거짓 판별 : if v
        return bool(abs(self))
    
    # 덧셈
    def __add__(self, other):   # '+' 연산자를 사용할 경우 호출되는 메서드
        x = self.x + other.x
        y = self.y + other.y
        return Vector(x, y)
    
    # 곱셈
    def __mul__(self, scalar):   # '*' 연산자를 사용할 때 호출되는 메서드
        return Vector(self.x * scalar, self.y * scalar)
```

```python
v = Vector(3, 4)
abs(v)
```

```python
# 객체를 일반 연산자를 이용하여 계산 가능

v1 = Vector(1, 2)
v2 = Vector(2, 4)
v1 + v2
```

### "객체를 숫자처럼 다룬다"는 말의 의미,
- 파이썬의 연산자와 함수들이 객체에도 그대로 적용되도록 만든다는 의미
- +, *, abs(), if, == 같은 문법이 직접 만든 객체에서도 
동일한 개념으로 동작하게 만든다는 뜻

```python
# 일반 숫자에서 
a = 3
b = 4
a + b
```

```python
# 벡터 객체에서
v1 = Vector(3, 4)
v2 = Vector(1, 2)
v1 + v2      # 객체의 + 연산이 가능하다 : 문법은 동일하지만 직접 정의한 로직에 의한 결과
```

```python
# 실제 연산은 다음과 같다
v1.__add__(v2)

# 연산자 : 특별 메서드 호출의 syntactic sugar(문법적 설탕)
# 훨씬 직관적이고 수학처럼 읽히는 코드로 작성 가능하다.

# "객체를 숫자처럼 다룬다"는 것은
# 파이썬의 연산자와 기본 문법을 객체에 연결해서,
# 코드를 수학 공식처럼 읽히기 만드는 것
```

```python
# 실무적인 예시 : 그래픽 / UI 개발
button_pos = Vector(100, 200)
offset = Vector(10, -5)

final_pos = button_pos + offset  # UI 요소 이동
final_pos
```
