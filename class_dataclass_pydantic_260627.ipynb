```python
# 일반 클래스와 dataclass의 차이점

# 반드시 일반 클래스를 사용해야 하는 경우 (260627)

# : 객체가 "데이터"보다 "행동(behavior)" 중심인 경우

class BankAccount:
    def __init__(self, owner: str, balance: int):
        self.owener = owner
        self.balance = balance

    def deposit(self, amount: int):
        self.deposit += amount

    def withdraw(self, amount: int):
        if amount > self.balance:     # 잔액이 인출 금액보다 작은 경우 행동
            raise ValueError("잔액 부족")
        self.balance -= amount

# BankAcount 객체의 핵심은 데이터를 저장하는 것이 아니라,
# 입금, 출금, 이체, 이자 계산 등의 "행동"과 비즈니스 규칙 등이다.
# 이런 객체는 dataclass 보다는 일반 클래스가 자유롭다.
```

```python
# 일반 클래스

class UserRequest:
    def __init__(self, username: str, password: str, age: int):
        self.username = username
        self.password = password
        self.age = age

# 객체 생성하기
user = UserRequest(
    username = "terry",
    password="1234",
    age=20
)

print(user.username)
print(user.password)
print(user.age)
```

```python
print(user)
```

```python
# 클래스는 값(value)이 아니라 메모리 주소를 비교한다.abs

user1 = UserRequest("terry", "1234", 30)
user2 = UserRequest("terry", "1234", 30)

print(user1 == user2)

# 편의 기능을 직접 구현하여,
# def __eq__(self, other): 
# 값이 동일하면 True가 출력되도록 한다. (260627)
```
