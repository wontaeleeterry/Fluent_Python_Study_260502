### Keyword Class Patterns
- 데이터클래스를 사용하여 키워드 기반으로 구조화된 데이터 객체를 생성하고 관리하는 디자인 패턴
- 핵심 특징,
- 데이터 중심 : 주로 데이터를 저장하고 전달하는데 목적
- 키워드 기반 : 키-값 쌍으로 데이터를 구조화
- 간단한 구조 : @dataclass 데코레이터를 활용하여 간결하게 정의
- 유연한 확장 : 메서드를 추가하여 데이터를 처리하고 조작 가능

```python
from dataclasses import dataclass, field

@dataclass
class Country:
    name: str
    code: str = ""
    cities: List[str] = field(default_factory=list)

    # 이 패턴은 국가, 도시 등의 데이터를 구조적으로 관리할 때 효과적
```

```python
# field() 함수는 dataclasses 모듈에서 데이터클래스의 필드를 
# 더 세밀하게 제어할 때 사용하는 함수 

# 1. 기본값 설정 (default_factory)

from dataclasses import dataclass, field

@dataclass
class ClubMember:
    name: str
    guest: list = field(default_factory=list)  # 매번 새로운 빈 리스트 생성


@dataclass
class Person:
    name: str
    age: int = field(default=0)    # 기본값을 설정
    email: str = field(default_factory=str)  # 기본값 생성 함수
    hobbies: list = field(default_factory=list)  # 가변 객체 처리

    # 필드를 비교 대상에서 제외
    password: str = field(default_factory=list, compare=False)

    # _repr_에서 제외
    secret: str = field(default_factory=list, repr=False)
```

### Keyword Class Patterns에 대한 예제
- 국가명, 도시이름, 그리고 국가 코드를 사용
- 데이터 클래스의 장점 :
- @dataclass 데코레이터를 사용하여 간단한 코드로 객체 생성 가능
- 자동으로 __init__, __repr__, __eq__ 메서드 생성
- 타입 힌트 지원

- 확장성 :
- add_city() 메서드를 통해 도시 추가 가능
- get_city_count() 메서드로 도시 개수 확인
- get_all_info() 메서드로 전체 정보 출력


```python
from dataclasses import dataclass, field
from typing import List

@dataclass
class Country:
    name: str
    cities: List[str] = field(default_factory=list)
    code: str = ""
    
    def add_city(self, city_name: str):
        """도시 추가 메서드"""
        self.cities.append(city_name)
    
    def get_city_count(self) -> int:
        """도시 개수 반환"""
        return len(self.cities)

# 국가 정보를 저장하는 데이터 클래스
@dataclass
class CountryData:
    country_name: str
    capital_city: str = ""
    country_code: str = ""
    cities: List[str] = field(default_factory=list)
    
    def __post_init__(self):
        """초기화 후 처리"""
        if not self.country_code:
            # 코드가 없으면 국가 이름을 기반으로 생성
            self.country_code = self.country_name[:3].upper()
    
    def add_city(self, city: str):
        """도시 추가"""
        self.cities.append(city)
    
    def get_all_info(self) -> str:
        """모든 정보 출력"""
        return f"국가: {self.country_name}, 코드: {self.country_code}, 수도: {self.capital_city}"
    
    def get_city_count(self) -> int:
        """도시 개수 반환"""
        return len(self.cities)

# 예제 사용
print("=== 국가 데이터 클래스 패턴 예제 ===")

# 1. 기본 국가 생성
korea = CountryData("Korea", "Seoul", "KR")
print(f"1. {korea.get_all_info()}")

# 2. 도시 추가
korea.add_city("Busan")
korea.add_city("Jeju")
print(f"2. 도시 목록: {korea.cities}")

# 3. 국가 정보 업데이트
japan = CountryData("Japan", "Tokyo", "JP")
print(f"3. {japan.get_all_info()}")

# 4. 코드가 없는 경우 자동 생성
france = CountryData("France", "Paris")
print(f"4. France 코드: {france.country_code}")

# 5. 빈 리스트를 사용한 예제
empty_country = CountryData("Empty Country")
print(f"5. 빈 국가 정보: {empty_country}")

# 6. 도시 개수 확인
print(f"6. 한국의 도시 수: {korea.get_city_count()}")

# 7. 데이터 클래스의 장점 보여주기
@dataclass
class CityInfo:
    name: str
    country_code: str = ""
    population: int = 0
    
    def get_full_name(self) -> str:
        return f"{self.name} ({self.country_code})"

# 도시 정보
seoul = CityInfo("Seoul", "KR", 9500000)
tokyo = CityInfo("Tokyo", "JP", 13900000)

print(f"7. {seoul.get_full_name()}")
print(f"   {tokyo.get_full_name()}")

# 8. 도시 리스트 관리
@dataclass
class CountryWithCities:
    name: str
    code: str = ""
    cities: List[CityInfo] = field(default_factory=list)
    
    def add_city(self, city_info: CityInfo):
        self.cities.append(city_info)
    
    def get_cities_by_country(self, country_code: str) -> List[CityInfo]:
        return [city for city in self.cities if city.country_code == country_code]

# 국가와 도시 관리
asia_countries = CountryWithCities("Asia")
seoul_info = CityInfo("Seoul", "KR", 9500000)
tokyo_info = CityInfo("Tokyo", "JP", 13900000)

asia_countries.add_city(seoul_info)
asia_countries.add_city(tokyo_info)

print(f"8. 아시아 국가의 도시:")
for city in asia_countries.cities:
    print(f"   {city.get_full_name()}")
```
