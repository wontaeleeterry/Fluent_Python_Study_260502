### 일급 객체(First-class Object)
- 다음 3가지 조건을 모두 충족할 때, 일급 객체라고 부른다
- 쉽게 말해 "변수에 담을 수 있는 일반적인 데이터(숫자나 문자열 등)와 완전히 동일한 대우를 받는 존재 뜻

- 1) 변수나 데이터 구조(리스트, 딕셔너리 등)에 할당할 수 있다.
- 2) 함수의 인자(Argument)로 전달할 수 있다.
- 3) 함수의 반환값(Return value)으로 사용할 수 있다.

- 일반적인 숫자 10이나 문자열 Hello는 당연히 이 3가지가 가능하므로 일급 객체이다.
- 파이썬에서 함수(function)역시 이 3가지를 완벽하게 지원한다.
#### 일급 객체로써 함수
- 함수를 오브젝트(객체)처럼 다룬다
- 함수이름 뒤에 ()를 붙여 실행하는 것 외에도, 함수 이름 그 자체를 일반 데이터처럼
- 이리저리 던지고 주무를 수 있다는 뜻


```python
# 함수를 변수에 할당하기

def hello():
    return "안녕하세요!"

# 함수를 실행한 '결과'를 담는 것이 아니라, '함수 자체'를 변수에 담습니다.
greet = hello  

print(greet())  # 출력: 안녕하세요!
```

```python
# 함수를 다른 함수의 인자로 전달하기

def call_twice(func):
    # 인자로 받은 함수(func)를 두 번 실행하는 함수
    func()
    func()

def say_hi():
    print("Hi!")

# call_twice 함수에 say_hi 함수를 인자로 전달합니다.
call_twice(say_hi)
# 출력:
# Hi!
# Hi!
```

```python
# 함수를 반환값으로 사용하기

def factory(power):
    # 함수 내부에서 새로운 함수를 정의합니다.
    def exponent(base):
        return base ** power
    return exponent  # 함수 자체를 반환합니다.

square = factory(2)  # 제곱을 해주는 함수를 반환받음
cube = factory(3)    # 세제곱을 해주는 함수를 반환받음

print(square(4))  # 4의 제곱 -> 출력: 16
print(cube(4))    # 4의 세제곱 -> 출력: 64
```

```python
# 팩토리얼 함수

# 1. 일반적인 재귀 방식의 팩토리얼 함수 정의
def factorial(n):
    return 1 if n <= 1 else n * factorial(n - 1)    # 재귀함수 (260605)

# 2. 다른 연산 함수들 정의
def square(n):
    return n ** 2

def cube(n):
    return n ** 3

# 3. 함수들을 일반 객체처럼 리스트에 담습니다!
math_functions = [factorial, square, cube]

# 4. 리스트를 돌며 각 함수를 실행합니다.
number = 5
for func in math_functions:
    # func.__name__을 통해 함수 객체의 이름에도 접근할 수 있습니다.
    print(f"{func.__name__}({number}) = {func(number)}")

# 출력:
# factorial(5) = 120
# square(5) = 25
# cube(5) = 125
```

```python
# 파이썬 데코레이터 예제: 자동 함수 등록기

# 1. 비어있는 registry 리스트 생성
registry = []

# 2. Wrapper(데코레이터) 정의
def register(func):
    """
    이 함수가 바로 'Wrapper' 역할을 합니다.
    전달받은 함수(func)를 리스트에 등록하고, 그 함수를 그대로 반환합니다.
    """
    registry.append(func)
    return func

# 3. 여러 메서드 정의 (일부는 @register를 사용, 일부는 미사용)

@register
def say_hello():
    print("안녕하세요!")

@register
def say_goodbye():
    print("안녕히 가세요!")

def say_hidden():
    print("저는 등록되지 않은 비밀 함수입니다.")

# --- 실행 및 검증 ---

print(f"현재 등록된 함수 개수: {len(registry)}개")
print("-" * 30)

# 등록된 함수들만 순회하며 실행하기
print("등록된 함수들을 실행합니다:")
for function in registry:
    print(f"실행 중: {function.__name__}")
    function()
```

```python
print(register.__doc__)   # docstring 출력 함수 (260605)
```
