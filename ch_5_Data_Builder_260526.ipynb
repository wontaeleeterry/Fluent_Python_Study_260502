### 데이터 클래스 빌더
- 데이터 클래스 빌더는 필수적인 : __init__(), __repr__(), __eq__() 메서드를 자동으로 구현해주며,
- 여러 유용한 기능을 제공한다. (어떤기능?)

```python
# 예제 : 지리적 좌표 쌍을 나타내는 간단한 클래스 구현

# 파일로 저장해둔다면, coordinate.py
class Coordinate:
    def __init__(self, lat, lon):
        self.lat = lat
        self.lon = lon
```

```python
# 파일로 저장했다면,
# from coordinate import Coordinate

moscow = Coordinate(55.76, 37.62)
moscow

# __repr__() 이 없다?? : 아래 출력된 정보는 아이디만 출력함
```

```python
location = Coordinate(55.76, 37.62)
location == moscow   

# __eq__() 메서드는 객체의 ID를 비교하므로, False로 출력됨
```

```python
# 정확한 값을 비교한다면, 결과는 True로 출력됨

(location.lat, location.lon) == (moscow.lat, moscow.lon)
```

### dataclass

```python
# dataclass

from dataclasses import dataclass

@dataclass(frozen=True)
class Coordinate:
    lat: float
    lon: float
    
    def __str__(self):
        ns = 'N' if self.lat >= 0 else 'S'
        we = 'E' if self.lon >= 0 else 'W'
        return f'{abs(self.lat):.1f} degree {ns}, {abs(self.lon):.1f} degree {we}'
```

```python
moscow = Coordinate(55.76, 37.62)
print(moscow)
# __str__() : 문자열로 가공하여 결과 출력
```

```python
moscow   # __repr__() 호출 결과 출력
```

```python
location = Coordinate(55.76, 37.62)
moscow == location                     
# __eq__() 정상적으로 작동
```

```python
# 잘못된 예제

from dataclasses import dataclass

@dataclass
class ClubMember:
    name: str
    guest: list = []  # 이렇게 하면 안 됨 (260528)

# 두 개의 인스턴스를 생성
member1 = ClubMember("Alice")
member2 = ClubMember("Bob")

# 오류가 발생하는 이유는 
# Python에서 가변 기본 인수(mutable default argument)를 사용했기 때문입니다.
# guest: list = []와 같이 정의할 때, 빈 리스트 []는 클래스가 정의될 때 단 한 번 생성되며, 
# 각 인스턴스를 생성할 때마다 새로 만들어지지 않습니다. 
# 따라서 모든 ClubMember 인스턴스가 같은 리스트 객체를 공유하게 됩니다.
```

```python
# field(defaul_factory=list)

# default_factory는 dataclasses 모듈에서 제공하는 특별한 파라미터로, 
# 데이터 클래스의 필드에 대한 기본값을 생성하는 함수를 지정할 수 있게 해줍니다.

from dataclasses import dataclass, field

@dataclass
class ClubMember:
    name: str
    guest: list = field(default_factory=list)

# 작동 방식
# 각 인스턴스마다 새로운 객체 생성: default_factory=list를 사용하면, 
# 각 인스턴스를 만들 때마다 새로운 빈 리스트가 생성됩니다.
# 공유되지 않음:
```

```python
member1 = ClubMember("Alice")
member2 = ClubMember("Bob")

member1.guest.append("Charlie")
print(member2.guest)  # [] - member2는 영향을 받지 않음
```
