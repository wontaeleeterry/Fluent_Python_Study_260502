### Mapping vs dict 차이

| 구분         | dict   | Mapping           |
| ---------- | ------ | ----------------- |
| 실제 데이터 저장소 | O      | X                 |
| 인터페이스(규칙)  | X      | O                 |
| 바로 사용 가능   | O      | 보통 상속해서 사용        |
| 수정 가능      | O      | 기본은 읽기 전용         |
| 목적         | 데이터 저장 | dict처럼 동작하는 객체 정의 |


```python
# dict = 실제 사전
# Mapping = "사전처럼 동작해야 한다"는 규칙


# dict는 실제 데이터 구조
d = {
    "name": "Tom",
    "age": 20
}
d
```

```python
# Mapping은 규칙(인터페이스)

# Mapping 자체는 데이터를 저장하지 않습니다.
# 대신:
# "이 클래스는 dict처럼 행동할 것이다" 를 정의합니다.

# dict = 실제 자동차 (완성품) : 바로 사용
# Mapping = 자동차 설계 규칙 (규격) : 상속해서 사용

from collections.abc import Mapping

class Config(Mapping):

    def __init__(self, data):
        self._data = data

    def __getitem__(self, key):
        return self._data[key]

    def __iter__(self):
        return iter(self._data)

    def __len__(self):
        return len(self._data)

```

```python
config = Config( {
    "host": "localhost" # dict로 저장
})

print(config.keys())
```

```python
for key, value in config:
    print(key, value)
```

```python
from collections.abc import Mapping

class Config(Mapping):

    def __init__(self, data):
        self._data = dict(data)   # 입력받은 list를 dict로 저장 (260518)

    def __getitem__(self, key):
        return self._data[key]

    def __iter__(self):
        return iter(self._data)

    def __len__(self):
        return len(self._data)



# 키 - 값 쌍의 리스트를 정의해두고 매핑을 한다면,
# Mapping class 수정이 필요하다.

config_data = [("host", "localhost"),
               ("port", 3306),
            ]

config = Config(config_data)
config
```

```python
print(config["host"])

for key in config:
    print(key, config[key])
```

```python
for key, value in config.items():
    print(key, value)
```
