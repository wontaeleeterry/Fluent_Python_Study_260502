```python
# Annotation

# 파이썬에서 "어노테이트(annotate)한다"는 말은 코드에 추가적인 정보를 붙여주는 것을 의미합니다. 
# 이 정보는 프로그램의 실행에 직접적인 영향을 주기보다는 사람이나 개발 도구(IDE, 타입 체커 등)가 코드를 더 잘 이해하도록 돕기 위한 메타데이터입니다.
```

```python
# 어노테이션은 파이썬 인터프리터가 사용하는 정보가 아니라 개발자를 위한 정보입니다.

# 함수에서도 사용가능

def add(a: int, b: int) -> int:
    return a + b

# add 함수는 int 두 개를 받아서 int를 반환한다.

add("a", "b")  # 문자열을 넣어도 일단 실행은 된다.
```

```python
# 어노테이션이 없는 경우,

def add(a, b):
    return a + b

# 이 함수는
# int float str list
# 모두 받을 수 있습니다.

# 어노테이션이 있을 경우,
# IDE가 -> "문자열을 넣으려고 하는데?"
# 라고 경고해 줄 수 있습니다.
```

```python
# 실행 시 타입을 강제하는 방법
# beartype, typeguard

from beartype import beartype

@beartype
def add(a: int, b: int) -> int:
    return a + b
```

```python
add("a", "b")    #beartype 데코레이터로 자동 검사
```

```python
add(1, 2)
```

```python
# 타입 힌트가 없는 상테에서 beartype 데코레이터를 사용하면 (260614)

from beartype import beartype

@beartype
def add(a, b):
    return a + b  
```

```python
add("a", "b")    # 아무 일도 일어나지 않는다 (260614)
```
