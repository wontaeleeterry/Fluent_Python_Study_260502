### 파이썬 다운 카드 한 벌 (deck) : pp. ~49 (260502)

```python
import collections

Card = collections.namedtuple('Card', ['rank', 'suit'])   # 카드 한 장을 표현하는 자료형 정의

class FrenchDeck:
    ranks = [str(n) for n in range(2, 11)] + list('JQKA')    # 리스트 정의 : 리스트 컴프리헨션
    suits = 'spades diamonds clubs hearts'.split()

    def __init__(self):
        self._cards = [Card(rank, suit) for suit in self.suits for rank in self.ranks]
        # 리스트 컴프리헨션 : 모든 무늬 (4) x 모든 숫자 (13) 의 조합 생성 = 52장 생성
        # _cards : _는 내부용 변수라는 의미(관례)

    def __len__(self):
        return len(self._cards)   # 카드 개수 반환 : len(deck)이 가능해짐

    def __getitem__(self, position):   # __getitem__ : 카드 한 벌(deck)에서 순서를 쉽게 확인 가능하다.
        return self._cards[position]
        # 카드 꺼내기 : 리스트처럼 사용 가능해짐 : deck[0]
```

```python
# 일련의 속성으로만 정의된 간단한 클래스를 만들어 
# 한 장의 카드를 정의

beer_card = Card('7', 'diamond')
beer_card
```

```python
# 카드 한 벌(deck)을 표현하는 방법

deck = FrenchDeck()
len(deck)
```

### 정렬
- 카드의 순위를 정하는 함수 정의
- 클로버 2 : 0
- 스페이스 A : 51

```python
# 무늬 (suit) 우선 순위 정하기
suit_values = dict(spades=3, hearts=2, diamonds=1, clubs=0)

# 카드 하나를 입력받아 숫자 점수 하나를 반환
# FrenchDeck.ranks = ['2', '3', '4', ... ,'K', 'A'] : FrenchDeck.ranks.index('2') #0
def spades_high(card):
    rank_value = FrenchDeck.ranks.index(card.rank)
    return rank_value * len(suit_values) + suit_values[card.suit]  
    # 숫자 점수 * 4 + 무늬 점수 
    # 왜 4를 곱하는가 : 숫자가 올라가면 점수가 확 커짐
```

```python
# 함수 호출, 함수 전달의 의미 차이

for card in sorted(deck, key=spades_high):  # spades_high 메서드를 나중에 쓰라는 의미, 각 카드마다 호출
    print(card)

# sorted : deck 안에 카드들을 spades_high(card) 기준으로 정렬해서 새 리스트를 반환
# deck (원본)은 그대로 유지되고 정렬된 새 리스트가 생성됨
# sorted([3, 1, 2]) -> [1, 2, 3] : 새 리스트 생성
# key = spades_high : 각 요소를 비교할 때 이 함수를 기준으로 정렬하라
# 각각의 점수를 계산하고 정렬 수행
```
