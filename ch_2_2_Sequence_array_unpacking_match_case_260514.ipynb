```python
# 예제: 중첩 튜플을 구조 분해하고 match case로 필터링

# 데이터 준비
capitals = [
    ("서울", "KR", (37.566, 126.978)),
    ("도쿄", "JP", (35.689, 139.692)),
    ("파리", "FR", (48.857, 2.352)),
    ("뉴욕", "US", (40.713, -74.006)),
    ("런던", "GB", (51.507, -0.128)),
    ("시드니", "AU", (-33.868, 151.209))
]

# 구조 분해와 match case를 이용한 처리
for capital in capitals:
    # 구조 분해
    city, country_code, (latitude, longitude) = capital
    
    # match case로 위도가 양수인 경우만 출력
    match latitude:
        case lat if lat > 0:
            print(f"수도: {city}, 위도: {latitude}, 경도: {longitude}")
        case _:
            # 음수인 경우는 무시
            pass

'''
이 코드는 다음과 같은 방식으로 작동합니다:

capitals 튜플 리스트를 정의하고, 각 요소는 (수도 이름, 국가 코드, (위도, 경도))로 구성됩니다.
for 루프를 사용하여 각 튜플을 구조 분해합니다:
city: 수도 이름
country_code: 국가 코드
(latitude, longitude): 위도와 경도로 구성된 중첩 튜플
match case 문을 사용하여 위도가 양수인 경우만 출력합니다:
case lat if lat > 0: 조건을 만족하는 경우에만 출력
음수인 경우는 무시됩니다
결과로 위도가 양수인 수도만 출력됩니다 (서울, 도쿄, 파리, 시드니).
'''
```

```python
# 실무 사례 : * 사용하기

## 데이터 정제 - 헤더와 실제 데이터 분리

def clean_data(raw_data):
    """데이터에서 헤더와 실제 데이터 분리"""
    if not raw_data:
        return [], []

    # 헤더(1개)와 데이터(나머지) 분리
    header, *data = raw_data

    # 데이터 정제 (공백 제거)
    cleaned_data = [item.strip() for item in data if item.strip()]

    return header, cleaned_data

# 사용 예 : 날짜 그리고 공백은 삭제됨 (260514)
raw = ["날짜", "2023-01-01", "2023-01-02", "2023-01-03", ""]
header, clean_data = clean_data(raw)
print(f"헤더: {header}")  # 헤더: 날짜
print(f"데이터: {clean_data}")  # 데이터: ['2023-01-01', '2023-01-02', '2023-01-03']
```

```python
# API 응답에서 중요한 정보만 추출
def parse_api_response(response_data):
    """API 응답에서 메타데이터와 본문 분리"""
    status, message, *data = response_data   # 나머지 데이터 부분만 추출
    return {                
        "status": status,
        "message": message,
        "data": data        # 딕셔너리 형태로 반환
    } 

# 사용 예
api_response = ["success", "데이터 조회 성공", {"id": 1}, {"id": 2}, {"id": 3}]

# 정의한 함수를 이용하여 api 응답을 분석
parsed = parse_api_response(api_response)
print(f"상태: {parsed['status']}")
print(f"데이터: {parsed['data']}")

# 출력:
# 상태: success
# 데이터: [{'id': 1}, {'id': 2}, {'id': 3}]
```

```python
## 함수 매개변수에서 사용 : 가변인자 전달

def process_scores_better(*scores):
    """가장 높은 점수와 평균 점수를 계산"""
    if not scores:
        return 0, 0

    highest = max(scores)
    average = sum(scores) / len(scores)
    return highest, average

scores = (85, 92, 78, 96, 88, 91)    # 리스트, 튜플 둘다 가능하다 (260514)

# 실제 사용 예시
print("scores의 최고점:", process_scores_better(*scores))

'''
#### `*scores`의 역할:

1. **리스트 풀기**: `[85, 92, 78, 96, 88, 91]` → `85, 92, 78, 96, 88, 91`
2. **가변 인자 전달**: `process_scores_better(*scores)`는 `process_scores_better(85,
92, 78, 96, 88, 91)`과 동일
3. **함수에 여러 값을 하나씩 전달**: 함수 내부에서 `*scores`는 `(85, 92, 78, 96, 88,
91)` 형태로 받게 됩니다

**결론**: `*scores`는 리스트를 함수의 각 인자로 분해하여 전달하는 역할을 합니다.
'''
```

```python
# 가상의 로봇팔의 움직임을 컨트롤하는 클래스의 메서드를 파이썬으로 작성한다. 
# match, case를 이용하여 예시를 작성해주세요.

class RoboticArm:
    def __init__(self):
        self.position = [0, 0, 0]  # [x, y, z] 좌표
        self.angle = [0, 0, 0]     # [shoulder, elbow, wrist] 각도
        self.gripper_open = False

    def move_to_position(self, x, y, z):
        """지정된 위치로 이동"""
        self.position = [x, y, z]
        print(f"로봇팔을 위치 ({x}, {y}, {z})로 이동")

    def rotate_joint(self, joint_index, angle):
        """지정된 관절 회전"""
        if 0 <= joint_index < 3:
            self.angle[joint_index] = angle
            print(f"관절 {joint_index}을(를) {angle}도로 회전")

    def open_gripper(self):
        """그리퍼 열기"""
        self.gripper_open = True
        print("그리퍼 열기")

    def close_gripper(self):
        """그리퍼 닫기"""
        self.gripper_open = False
        print("그리퍼 닫기")

    def execute_command(self, command):
        """명령어에 따라 로봇팔 동작 실행"""
        match command:
            case "home":
                self.position = [0, 0, 0]
                self.angle = [0, 0, 0]
                self.gripper_open = False
                print("로봇팔을 홈 위치로 이동")

            case "pickup":
                self.move_to_position(10, 15, 5)
                self.rotate_joint(0, 45)
                self.rotate_joint(1, 30)
                self.open_gripper()
                print("물체 집기 준비 완료")

            case "place":
                self.move_to_position(20, 10, 8)
                self.rotate_joint(0, 60)
                self.close_gripper()
                print("물체 놓기 완료")

            case "scan":
                self.move_to_position(5, 5, 10)
                self.rotate_joint(2, 90)
                print("스캔 모드로 이동")

            case "reset":
                self.position = [0, 0, 0]
                self.angle = [0, 0, 0]
                self.gripper_open = False
                print("로봇팔 초기화 완료")

            case "hold":
                self.move_to_position(15, 20, 3)
                self.rotate_joint(0, 30)
                self.rotate_joint(1, 45)
                self.open_gripper()
                print("물체 유지 상태")

            case _:
                print(f"알 수 없는 명령어: {command}")

    def get_status(self):
        """로봇팔 상태 출력"""
        return {
            "position": self.position,
            "angle": self.angle,
            "gripper_open": self.gripper_open
        }

# 사용 예시
if __name__ == "__main__":
    # 로봇팔 인스턴스 생성
    arm = RoboticArm()

    # 다양한 명령어 실행
    commands = ["home", "pickup", "place", "scan", "hold", "unknown"]

    for cmd in commands:
        print(f"\n=== 명령어: {cmd} ===")
        arm.execute_command(cmd)

        # 상태 확인
        status = arm.get_status()
        print(f"현재 상태: {status}")
        print("-----------------------------------------")
```

```python
# 시퀀스를 이해하기 위한 실무 예제

class OrderProcessor:
    def __init__(self):
        # 시퀀스를 활용한 주문 관리
        self.orders = []
    
    def add_order(self, order_items):
        # 시퀀스로 주문 항목 관리
        self.orders.append(order_items)
    
    def process_orders(self):
        # 시퀀스 순서대로 주문 처리
        processed_orders = []
        for order in self.orders:
            # 시퀀스 요소에 대한 처리
            processed_order = [item.upper() for item in order] # 대문자 처리
            processed_orders.append(processed_order)
        return processed_orders      # 대문자 처리된 결과를 반환
    
    def get_order_by_index(self, index):
        # 시퀀스 인덱싱을 통한 접근
        return self.orders[index] if 0 <= index < len(self.orders) else None

# 실무 사례 실행
processor = OrderProcessor()

# 주문 항목 시퀀스 추가
processor.add_order(["apple", "banana", "orange"])
processor.add_order(["cherry", "date", "elderberry"])

# 시퀀스 순서대로 처리
processed = processor.process_orders()
print("Processed Orders:", processed)

# 시퀀스 인덱싱으로 접근
first_order = processor.get_order_by_index(0)   # 0 : 첫번째 주문을 반환
print("First Order:", first_order)
```

```python
import random

# 3개의 리스트로 구성된 틱택토 보드를 생성
# 각 리스트는 3개의 요소로 구성되며, 'X', 'O', 또는 빈 공간(' ')으로 표현
board = [
    [' ', ' ', ' '],  # 첫 번째 행: 빈 공간으로 초기화
    [' ', ' ', ' '],  # 두 번째 행: 빈 공간으로 초기화
    [' ', ' ', ' ']   # 세 번째 행: 빈 공간으로 초기화
]

# 게임 상태를 나타내는 변수
game_over = False  # 게임 종료 여부

# 현재 플레이어를 나타내는 변수 ('X' 또는 'O')
current_player = 'X'

# 게임 보드 출력 함수
def print_board():
    """틱택토 보드를 화면에 출력하는 함수"""
    print("-------------")
    for i in range(3):
        print(f"| {board[i][0]} | {board[i][1]} | {board[i][2]} |")
        print("-------------")

# 승리 조건 확인 함수
def check_winner():
    """틱택토 게임에서 승자 확인"""

    # 가로 방향으로 승리 조건 검사
    for i in range(3):
        if board[i][0] == board[i][1] == board[i][2] != ' ':
            return board[i][0]

    # 세로 방향으로 승리 조건 검사
    for i in range(3):
        if board[0][i] == board[1][i] == board[2][i] != ' ':
            return board[0][i]

    # 대각선 방향으로 승리 조건 검사
    if board[0][0] == board[1][1] == board[2][2] != ' ':
        return board[0][0]
    if board[0][2] == board[1][1] == board[2][0] != ' ':
        return board[0][2]

    # 승자가 없을 경우 None 반환
    return None

# 무승부 조건 확인 함수
def is_board_full():
    """보드가 모두 채워졌는지 확인"""
    for i in range(3):
        for j in range(3):
            if board[i][j] == ' ':
                return False
    return True

# 게임 메인 루프
print("틱택토 게임을 시작합니다!")
print_board()

# 게임 루프
while not game_over:
    print(f"현재 플레이어: {current_player}")

    # 플레이어의 입력을 받기 위한 반복문
    while True:
        try:
            # 플레이어에게 위치 입력 받기
            row = int(input("행을 입력하세요 (0-2): "))
            col = int(input("열을 입력하세요 (0-2): "))

            # 입력된 위치가 유효한지 확인
            if 0 <= row <= 2 and 0 <= col <= 2:
                # 해당 위치가 비어 있는지 확인
                if board[row][col] == ' ':
                    # 위치에 현재 플레이어의 마크를 설정
                    board[row][col] = current_player
                    break  # 유효한 입력이면 루프 종료
                else:
                    print("해당 위치는 이미 사용되었습니다. 다시 입력하세요.")
            else:
                print("행과 열은 0에서 2 사이의 숫자여야 합니다. 다시 입력하세요.")
        except ValueError:
            print("유효한 숫자를 입력하세요. 다시 입력하세요.")

    # 보드 출력
    print_board()

    # 승자 확인
    winner = check_winner()
    if winner:
        print(f"플레이어 {winner}가 승리했습니다!")
        game_over = True
    elif is_board_full():
        print("무승부입니다!")
        game_over = True
    else:
        # 다음 플레이어로 전환
        current_player = 'O' if current_player == 'X' else 'X'
```
